#Popularity Based Model

```python
#we'll merge both the rating and users on ISBN column, we'll show top 50 highest rated books
```

```python
ratings_with_name = ratings.merge(books, on='ISBN')
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns= {'Book-Rating':'num_ratings'},inplace= True)
num_rating_df.head(10)
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns= {'Book-Rating':'avg_ratings'},inplace= True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df =popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popular_df
```

```python
popular_df =popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

#Collaborative Filtering


```python
#collaborative filtering
#our criteria for choosing books is
# we will consider only the users who has  rated  for more than 200 books and we will select  only those books which has 50+ ratings
```

```python
x =ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
users_who_rate = x[x].index
```

```python
filtered_rating =ratings_with_name[ratings_with_name['User-ID'].isin(users_who_rate)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
famous_books.shape
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.shape
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt.fillna(0 ,inplace=True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity       #used to find distance of each point from all other points
```

```python
similarity_scores=cosine_similarity(pt)                      #returns a numpy array of all distances
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
  # index fetch
  index=np.where(pt.index==book_name)[0][0]
  distances=similarity_scores[index]
  similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x: x[1], reverse=True)[1:6]
  print("{:<70} {:<50}".format("Title of book","Similarity Score"))
  print()
  val=similar_items[0][1]
  for i in similar_items:
    print("{:<70} {:<50}".format(pt.index[i[0]], i[1]*100))
```

```python
recommend('Harry Potter and the Prisoner of Azkaban (Book 3)')
```
